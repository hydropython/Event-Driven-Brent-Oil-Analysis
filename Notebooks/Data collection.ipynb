```python
import pandas as pd
import pandas as pd

import pandas as pd

# Load datasets
exchange = pd.read_csv('../Data/Exchange rate.csv')
macroeconomic=pd.read_csv('../Data/macroeconomic .csv')
oil=pd.read_csv('../Data/Copy of BrentOilPrices.csv')
# Check unique values in the 'Date' column to confirm format
print(exchange['Date'].unique())

# Specify the date format when converting
exchange['Date'] = pd.to_datetime(exchange['Date'], format='%b-%y', errors='coerce')

# Check for any invalid dates
print(exchange[exchange['Date'].isna()])

# Drop or handle NaT values as needed
exchange.dropna(subset=['Date'], inplace=True)

# Convert 'Date' columns to datetime
exchange['Date'] = pd.to_datetime(exchange['Date'])
macroeconomic['Date'] = pd.to_datetime(macroeconomic['Date'])
oil['Date'] = pd.to_datetime(oil['Date'])

# Set 'Date' as index
exchange.set_index('Date', inplace=True)
macroeconomic.set_index('Date', inplace=True)
oil.set_index('Date', inplace=True)

# Resample macroeconomic data to monthly
macroeconomic_monthly = macroeconomic.resample('M').ffill()

# Merge datasets
combined_data = oil.join(exchange, how='outer').join(macroeconomic_monthly, how='outer')

# Fill missing values
combined_data.fillna(method='ffill', inplace=True)

# Inspect combined data
print(combined_data.head())


# Save the combined data to CSV
combined_data.to_csv('../Data/combined_data.csv')
```
